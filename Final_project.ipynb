```python
# load the dataset
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/heelafakher/Data-Scientist/main/data_Scientist_cleaned_2021.csv')
df
# find the correct datatype for each columns 
df.info
df.dtypes
```

```python
## finding for each columns of dataset mean, meadian and mode
mean_upper_salary = df['Upper Salary'].mean()
median_upper_salary = df['Upper Salary'].median()
mode_upper_salary = df['Upper Salary'].mode()
print("upper salary mean is :",mean_upper_salary )
print("upper salary median is :",median_upper_salary)
print("upper salary mode is:",mode_upper_salary )
```

```python
mean_lower_salary = df['Lower Salary'].mean()
median_lower_salary = df['Lower Salary'].median()
mode_lower_salary = df['Lower Salary'].mode()
print("lower salary mean is :",mean_lower_salary )
print("lower salary median is :", median_lower_salary)
print("lower Salary mode is:", mode_lower_salary )
```

```python
mean_r = df['Rating'].mean()
median_r= df['Rating'].median()
mode_r  = df['Rating'].mode()
print("Rating mean is :",mean_r)
print("Rating median is :",median_r)
print("Rating mode is:",mode_r)
```

```python
mean_avsalary = df['Avg Salary'].mean()
median_avsalary = df['Avg Salary'].median()
mode_avsalary = df['Avg Salary'].mode()
print("Avg salary mean is :",mean_avsalary)
print("Avg salary median is :",median_avsalary)
print("Avg Salary mode is:",mode_avsalary)
```

```python
mean_found = df['Founded'].mean()
median_found= df['Founded'].median()
mode_found = df['Founded'].mode()
print("founded mean is :",mean_found)
print("founded median is :",median_found)
print("founded mode is:",mode_found)
```

```python
#correlation of columns
df.corr()
```

```python
# Scatter plot of Lower salary vs. upper salary
plt.scatter(df['Upper Salary'],df['Lower Salary'], alpha=0.5)
plt.xlabel('Upper salary')
plt.ylabel('Lower salary')
plt.title('upper salary vs. lower salary')
plt.show()

# Correlation between upper salary and lower salary
correlation = df['Lower Salary'].corr(df['Upper Salary'])
print('Correlation between upper salary and lower salary:', correlation)
# Check if the correlation is significant
if abs(correlation) > 0.9:
    print('The correlation is significant')
else:
    print('The correlation is not significant')
```

```python
# Bar gragh of company rating :
df.sort_values(['Rating'],ascending= True,inplace=True)
print(df)
company_rating = df.groupby('Company Name')['Rating'].mean().head(10)
company_rating.plot(kind ='bar')
plt.xlabel('Company Name')
plt.ylabel('Rating')
plt.title('company rating')
plt.show()
######################################################
#bar gragh for which company has the highest rating:
highest_rating = company_rating.idxmax()
print('company name  with highest rating:',highest_rating)
###################################################
 #bar gragh for which company  has the lowest rating:
lowest_rating = company_rating.idxmin()
print('company name  with lowest rating:',lowest_rating)
```

```python
# Pie chart of avg salary by job title :  

avgsalary_by_jobtitle = df.groupby('Job Title')['Avg Salary'].sum().head(5)
print(avgsalary_by_jobtitle)
avgsalary_by_jobtitle.plot(kind = 'pie',autopct='%1.1f%%')
plt.title('avg salary via job title')
plt.show()
```

```python
# Bar ghragh of avarege salary Vs compnay type of owner ship:
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/heelafakher/Data-Scientist/main/data_Scientist_cleaned_2021.csv')
typ_of_ownership = df.groupby('Type of ownership')['Avg Salary'].sum().head(10)
typ_of_ownership.plot(kind ='bar')
plt.xlabel('Type of ownership')
plt.ylabel('Avg Salary')
plt.title('Avarage Salary VS type of company ownership')
plt.show()
######################################################
# showing the highest averge salary based of type of ownership:
highest_owner= typ_of_ownership.idxmax()
print('company ownership with highest average salary:',highest_owner)
###################################################
#showing the lowest averge salary based of type of ownership:
lowest_owner =typ_of_ownership.idxmin()
print('company ownerhsip with lowest average salary',lowest_owner)
```
